### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular ingresos totales por tienda
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

# Mostrar resultados de forma organizada
print(f"Ingreso Total Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Ingreso Total Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Ingreso Total Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Ingreso Total Tienda 4: ${ingreso_tienda4:,.2f}")
```

```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

plt.figure(figsize=(10, 6))
plt.bar(tiendas, ingresos, color=['skyblue', 'orchid','lightgreen', 'salmon' ])
plt.title('Comparativa de Ingresos Totales por Tienda')
plt.ylabel('Ingresos ($)')
plt.show()
```

# 2. Ventas por categoría

```python
# Contar ventas por categoría en cada tienda
ventas_cat_tienda1 = tienda['Categoría del Producto'].value_counts()
ventas_cat_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventas_cat_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventas_cat_tienda4 = tienda4['Categoría del Producto'].value_counts()

# Mostrar resultados de las categorías más populares
print("--- Ventas por Categoría: Tienda 1 ---")
print(ventas_cat_tienda1)
print("\n--- Ventas por Categoría: Tienda 2 ---")
print(ventas_cat_tienda2)
print("\n--- Ventas por Categoría: Tienda 3 ---")
print(ventas_cat_tienda3)
print("\n--- Ventas por Categoría: Tienda 4 ---")
print(ventas_cat_tienda4)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

ventas_cat_tienda1.plot(kind='pie', ax=axes[0,0], autopct='%1.1f%%', title='Categorías Tienda 1')
ventas_cat_tienda2.plot(kind='pie', ax=axes[0,1], autopct='%1.1f%%', title='Categorías Tienda 2')
ventas_cat_tienda3.plot(kind='pie', ax=axes[1,0], autopct='%1.1f%%', title='Categorías Tienda 3')
ventas_cat_tienda4.plot(kind='pie', ax=axes[1,1], autopct='%1.1f%%', title='Categorías Tienda 4')

plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calcular la calificación promedio por tienda
calificacion_tienda1 = tienda['Calificación'].mean()
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda4 = tienda4['Calificación'].mean()

# Mostrar resultados de satisfacción
print(f"Satisfacción promedio Tienda 1: {calificacion_tienda1:.2f} ")
print(f"Satisfacción promedio Tienda 2: {calificacion_tienda2:.2f} ")
print(f"Satisfacción promedio Tienda 3: {calificacion_tienda3:.2f} ")
print(f"Satisfacción promedio Tienda 4: {calificacion_tienda4:.2f} ")
```

```python
import matplotlib.pyplot as plt

# Datos para la comparativa
tiendas_nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones = [calificacion_tienda1, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4]

plt.figure(figsize=(10, 5))
plt.plot(tiendas_nombres, calificaciones, marker='o', linestyle='--', color='gold', markersize=10)
plt.title('Nivel de Satisfacción del Cliente por Tienda')
plt.ylabel('Calificación Promedio (1-5)')
plt.ylim(1, 5)
plt.grid(axis='y', alpha=0.3)
plt.show()
```
