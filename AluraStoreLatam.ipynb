### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular ingresos totales por tienda
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

# Mostrar resultados de forma organizada
print(f"Ingreso Total Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Ingreso Total Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Ingreso Total Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Ingreso Total Tienda 4: ${ingreso_tienda4:,.2f}")
```

```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

plt.figure(figsize=(10, 6))
plt.bar(tiendas, ingresos, color=['skyblue', 'orchid','lightgreen', 'salmon' ])
plt.title('Comparativa de Ingresos Totales por Tienda')
plt.ylabel('Ingresos ($)')
plt.show()
```

# 2. Ventas por categoría

```python
# Contar ventas por categoría en cada tienda
ventas_cat_tienda1 = tienda['Categoría del Producto'].value_counts()
ventas_cat_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventas_cat_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventas_cat_tienda4 = tienda4['Categoría del Producto'].value_counts()

# Mostrar resultados de las categorías más populares
print("--- Ventas por Categoría: Tienda 1 ---")
print(ventas_cat_tienda1)
print("\n--- Ventas por Categoría: Tienda 2 ---")
print(ventas_cat_tienda2)
print("\n--- Ventas por Categoría: Tienda 3 ---")
print(ventas_cat_tienda3)
print("\n--- Ventas por Categoría: Tienda 4 ---")
print(ventas_cat_tienda4)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

ventas_cat_tienda1.plot(kind='pie', ax=axes[0,0], autopct='%1.1f%%', title='Categorías Tienda 1')
ventas_cat_tienda2.plot(kind='pie', ax=axes[0,1], autopct='%1.1f%%', title='Categorías Tienda 2')
ventas_cat_tienda3.plot(kind='pie', ax=axes[1,0], autopct='%1.1f%%', title='Categorías Tienda 3')
ventas_cat_tienda4.plot(kind='pie', ax=axes[1,1], autopct='%1.1f%%', title='Categorías Tienda 4')

plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calcular la calificación promedio por tienda
calificacion_tienda1 = tienda['Calificación'].mean()
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda4 = tienda4['Calificación'].mean()

# Mostrar resultados de satisfacción
print(f"Satisfacción promedio Tienda 1: {calificacion_tienda1:.2f} ")
print(f"Satisfacción promedio Tienda 2: {calificacion_tienda2:.2f} ")
print(f"Satisfacción promedio Tienda 3: {calificacion_tienda3:.2f} ")
print(f"Satisfacción promedio Tienda 4: {calificacion_tienda4:.2f} ")
```

```python
import matplotlib.pyplot as plt

# Datos para la comparativa
tiendas_nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones = [calificacion_tienda1, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4]

plt.figure(figsize=(10, 5))
plt.plot(tiendas_nombres, calificaciones, marker='o', linestyle='--', color='gold', markersize=10)
plt.title('Nivel de Satisfacción del Cliente por Tienda')
plt.ylabel('Calificación Promedio (1-5)')
plt.ylim(1, 5)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

# 4. Productos más y menos vendidos

```python
def analizar_productos(df, nombre_tienda):
    conteo = df['Producto'].value_counts()
    top_5 = conteo.head(5)
    bottom_5 = conteo.tail(5)

    print(f"--- {nombre_tienda} ---")
    print("MÁS VENDIDOS:")
    print(top_5)
    print("\nMENOS VENDIDOS:")
    print(bottom_5)
    print("-" * 30 + "\n")
    return top_5, bottom_5

top1, bot1 = analizar_productos(tienda, "Tienda 1")
top2, bot2 = analizar_productos(tienda2, "Tienda 2")
top3, bot3 = analizar_productos(tienda3, "Tienda 3")
top4, bot4 = analizar_productos(tienda4, "Tienda 4")
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# Datos para iterar el gráfico
tops = [top1, top2, top3, top4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colores = ['#2ecc71', '#3498db', '#9b59b6', '#f1c40f']

for i, ax in enumerate(axes.flat):
    tops[i].sort_values().plot(kind='barh', ax=ax, color=colores[i])
    ax.set_title(f'Top 5 Productos más vendidos: {nombres[i]}')
    ax.set_xlabel('Cantidad vendida')

plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
# Calcular costo de envío promedio por tienda
envio_tienda1 = tienda['Costo de envío'].mean()
envio_tienda2 = tienda2['Costo de envío'].mean()
envio_tienda3 = tienda3['Costo de envío'].mean()
envio_tienda4 = tienda4['Costo de envío'].mean()

# Mostrar resultados
print(f"Costo de envío promedio Tienda 1: ${envio_tienda1:.2f}")
print(f"Costo de envío promedio Tienda 2: ${envio_tienda2:.2f}")
print(f"Costo de envío promedio Tienda 3: ${envio_tienda3:.2f}")
print(f"Costo de envío promedio Tienda 4: ${envio_tienda4:.2f}")
```

```python
import matplotlib.pyplot as plt

tiendas_logistica = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costos_envio = [envio_tienda1, envio_tienda2, envio_tienda3, envio_tienda4]

plt.figure(figsize=(10, 6))

plt.bar(tiendas_logistica, costos_envio, color=['#ff9999','#ff6666','#ff3333','#cc0000'])
plt.title('Costo de Envío Promedio por Tienda')
plt.ylabel('Costo de Envío ($)')
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Etiquetas valor sobre las barras
for i, v in enumerate(costos_envio):
    plt.text(i, v + 50, f"${v:.2f}", ha='center', fontweight='bold')

plt.show()
```
